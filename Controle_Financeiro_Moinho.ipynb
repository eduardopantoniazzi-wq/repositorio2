```python
# ── CÉLULA 3: Upload dos arquivos ────────────────────────────────────────────
#
# Faça upload de:
#   • google_credentials.json   (credencial do Google)
#   • bradesco*.pdf / *.xlsx    (extrato Bradesco)
#   • sicredi*.pdf  / *.xlsx    (extrato Sicredi)
#   • bb*.pdf       / *.xlsx    (extrato Banco do Brasil)
#   • receitas_despesas*.xlsx   (planilha de previsão)
#
from google.colab import files
import shutil, os

print('Selecione TODOS os arquivos de uma vez (Ctrl+clique para selecionar múltiplos):')
uploaded = files.upload()

for nome, conteudo in uploaded.items():
    nome_lower = nome.lower()
    if 'credentials' in nome_lower or nome_lower.endswith('.json'):
        destino = f'repo/credentials/{nome}'
        # Renomeia para o nome esperado pelo sistema
        with open('repo/credentials/google_credentials.json', 'wb') as f:
            f.write(conteudo)
        print(f'  ✔ {nome} → credentials/google_credentials.json')
    else:
        destino = f'repo/data/extratos/{nome}'
        with open(destino, 'wb') as f:
            f.write(conteudo)
        print(f'  ✔ {nome} → data/extratos/')

print('\n✔ Todos os arquivos enviados!')
```

```python
# ── CÉLULA 4: Executar análise e atualizar Google Sheets ─────────────────────
#
# Altere os meses abaixo se necessário
# Exemplos: --meses Junho    |    --meses Maio Junho    |    --meses Janeiro
#
MESES = 'Junho'

os.chdir('/content/repo')
!python atualizar.py --meses {MESES}
```
